# Generate Fractals

```python
from utils import *
import matplotlib.pyplot as plt
```

```python
def show_fractal(fractal, x_min, x_max, y_min, y_max):
    plt.figure(figsize=(5,5))
    plt.imshow(fractal,
               interpolation="bilinear",
               cmap='gray',  # plt.cm.hot,
               vmax=abs(fractal).max(),
               vmin=abs(fractal).min(),
               extent=[x_min, x_max, y_min, y_max])
    plt.axis('off')
    plt.show()
```

```python
x_min, x_max, y_min, y_max = -4, 4, -4, 4
fractal, x_min, x_max, y_min, y_max = generate_fractals("Mandelbrot", x_min, x_max, y_min, y_max)
show_fractal(fractal, x_min, x_max, y_min, y_max)
```

```python
x_min, x_max, y_min, y_max = -4, 4, -4, 4
fractal, x_min, x_max, y_min, y_max = generate_fractals("Burning_ship", x_min, x_max, y_min, y_max)
show_fractal(fractal, x_min, x_max, y_min, y_max)
```

```python
# take z at mendelbrot edge
z = complex(-0.4, 0.6)
x_min, x_max, y_min, y_max = -4, 4, -4, 4
fractal, x_min, x_max, y_min, y_max = generate_fractals("Julia", x_min, x_max, y_min, y_max, z)
show_fractal(fractal, x_min, x_max, y_min, y_max)
```

```python
for i in range(3,7):
    print("n =",str(i))
    x_min, x_max, y_min, y_max = -4, 4, -4, 4
    fractal, x_min, x_max, y_min, y_max = generate_fractals("Multibrot", x_min, x_max, y_min, y_max, n=i)
    show_fractal(fractal, x_min, x_max, y_min, y_max)
```
